### Using lists instead of appending in np arrays are faster

```python
np.random.seed = 1
start = timeit.default_timer()
# start execution timer

result_1 = np.empty(shape=[0, 49])
list_chance = []
list_n = []

for n in np.arange(2, 50):

    # number of people selected
    iterations = 10000
    # number of trials
    mean_of_unique_counts = []
    temp_1 = []

    b_days = np.random.randint(1, 365, iterations * n).reshape(iterations, n)
    # create an n-d array of "n" random numbers and "iterations" rows
    # we want to see if there are duplicates for ever row where rows are the number of iterations

    for i in b_days:

        mean_of_unique_counts.append(
            np.array(np.unique(i, return_counts=True))[1].mean()
        )

    for i in mean_of_unique_counts:
        if i != 1:
            temp_1.append(i)
    # looping through entire set for row mean values noe equal to 1
    #     num_of_dupes = np.array(temp_1).shape[0]
    num_of_dupes = np.array(temp_1).shape[0]
    # gives the number of iterations that are have duplicated values

    print("the number of identical birthdays", num_of_dupes, "out of", iterations)
    prob_dupe = num_of_dupes / iterations
    print("the chances of pair occuring", f"{prob_dupe*100:.2f}", "%")

    list_chance.append(prob_dupe)
    list_n.append(n)

stop = timeit.default_timer()
# stop execution timer
print("Execution Time: ", stop - start)
# print(num_arrivals_2)
```

```python
fig, ax = plt.subplots()
ax = sns.scatterplot(y=list_chance, x=list_n)

ax.set(
    title="p of getting same bday vs n",
    xlabel="n",
    ylabel="chance of having same birthdays",
)

plt.show
```
