```python
import matplotlib.pyplot as plt
import numpy as np

%load_ext lab_black
%matplotlib inline

iters = 1000
# the number of iterations

plt.tight_layout()

prob = []
flip = []
# using loop instead of mapping. Slower but easier to follow logicaly

for i in range(1, iters):
    head = 0
    tail = 0
    # set initilize state where number of head and tail = 0
    for j in range(i):
        if np.random.randint(0, 2) == 0:
            head += 1
        else:
            tail += 1
    # for each iteration, if the random integer generated between 0 and 1, if the result is 0 then then the result is a head. Divide the number of heads by total number of tosses
    k = head / (i)
    prob.append(k)
    flip.append(i)


fig, axes = plt.subplots(nrows=2, ncols=1)
plt.subplots_adjust(hspace=0.8)
# adjust horizontal space between subplots
axes[0].hist(prob, bins=100, label="Probality of Heads")
# plot a histogram with 100 bins
axes[0].set(
    title="Distribution of the Probability of Heads Happening",
    xlabel="Probability of Heads",
    ylabel="Frequency",
)
# set title, x axis label, and y axis label
axes[0].set_xlim(0.3, 0.7)
# set x axis between 0.3 and 0.7


axes[1].plot(flip, prob)
# plot a scatter plot
axes[1].plot(flip, np.ones(iters - 1) * 0.5, color="black")
# add a horizontal line at y=0.5
axes[1].set(
    title="Probability of Heads vs Number of Flips",
    xlabel="Number of Flips",
    ylabel="Probability of Heads",
)
# set title, x axis label, y axis label
axes[1].grid(True)
# show gridlines
plt.show()
```

```python
plt.tight_layout()
n = 5000
# choose a number for the amount of data in poisson distribution
pop_dist = np.random.poisson(2, size=n)
# randomly generate a Possion distribtuion with lamda = 2 and size =n

iters = 20000
# number of iterations for simulation
random_sample_size = 100
# the sample size to be included for sample mean

mean_rnd_chosen = []
for i in range(0, iters):
    sample = np.random.choice(pop_dist, random_sample_size).mean()
    # randomly choose {random_sample_size} numbers and return the mean of the array
    mean_rnd_chosen.append(sample)
    # append the result into a list

mean_rnd_chosen = np.array(mean_rnd_chosen)
# make the list into a numpy array
print(f"the mean of the randonly chosen sample mean is: {mean_rnd_chosen.mean():.3f}")
# print the mean of the list with 3 decimal spaces
plt.subplots()
plt.hist(
    mean_rnd_chosen,
    200,
    label='Distribution of the "mean" of the "means" of the Poission distribution set',
)
# plot a histogram for the mean of the means

plt.title('Distribution of the "mean" of the "means" of the Poission distribution set')
# set the plot title
plt.xlabel("Mean of the mean of the Poisson distribution")
# set the x axis label
plt.ylabel("Frequency of occurence")
# set the y axis label

plt.show()
```
