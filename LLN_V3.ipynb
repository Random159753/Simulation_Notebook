# Law of Large Numbers (`LLN`) illustration
This notebook contains very simple snippets for illustration of `LLN`

### Importing necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from numpy import mean
#import math
#from statistics import mean
sns.set_style("ticks")
%matplotlib inline
```

## Example 1: `LLN` for rolling a die
### In this example, we will roll a die several times and each time we calculate average of the experimental outcome . True mean of a die-roll is \$\mathbb{E}(X) = \mu = 3.5 \$ where \$X\$ is the experimental outcome. 

```python
fig, axes = plt.subplots(figsize=(15, 8))
axes.set(
    title="Average of the a die roll outcome",
    xlabel="Sample size",
    ylabel="Experimental outcome average",
)

max_sample_size = 2000
sample_sizes = np.arange(1,max_sample_size, 1)
means = np.array([mean(np.random.randint(1, high = 7, size = sample_s)) for sample_s in sample_sizes])

sns.lineplot(data=means, color = "black")
plt.axhline(y=3.5, linewidth=4, color = 'red')
```

## Example 2: `LLN` for random number between `0` and `1`
### In this example, we randomly select numbers between `0` and `1` and each time calculate average of the  outcomes. True mean of uniform random variable between `0` and `1` is \$\mathbb{E}(X) = \mu = 0.5 \$. 

```python
fig, axes = plt.subplots(figsize=(15, 8))
axes.set(
    title="Average of the a uniform between 0 and 1",
    xlabel="Sample size",
    ylabel="Random number",
)

max_sample_size = 2000
sample_sizes = np.arange(1,max_sample_size, 1)
means = np.array([mean(np.random.random(size = sample_s)) for sample_s in sample_sizes])

sns.lineplot(data=means, color = "black")
plt.axhline(y=0.5, linewidth=4, color = 'red')
```
