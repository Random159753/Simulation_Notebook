```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math

sns.set_palette("Dark2")
sns.set_style("whitegrid", {"axes.grid": False})

%load_ext lab_black
%matplotlib inline
```

```python
dist_size = 5000
# choose a number for the amount of data in distribution
iters = 2000
# number of iterations for simulation
pop_dist = np.random.uniform(size=dist_size)
# randomly generate a uniform distribution
bin_num = int(1 + 3.322 * math.log(iters, 10))
# choosing number of bins
```

```python
def rnd_mean(n_c):
    a = np.random.choice(pop_dist, n_c).mean()
    return a


# define a function to calculate the mean of a random sample
```

```python
# consctructing arrays for randomly choosing sample size:
m2 = np.ones(iters, dtype=int) * 2
# array of 2 with "iters" number of items
m3 = np.ones(iters, dtype=int) * 5
# array of 5 with "iters" number of items
m4 = np.ones(iters, dtype=int) * 15
# array of 15 with "iters" number of items
m5 = np.ones(iters, dtype=int) * 30
# array of 30 with "iters" number of items
list_2 = []
list_3 = []
list_4 = []
list_5 = []
list_2.append(list(map(rnd_mean, m2)))
list_3.append(list(map(rnd_mean, m3)))
list_4.append(list(map(rnd_mean, m4)))
list_5.append(list(map(rnd_mean, m5)))
```

```python
def cons_array():
    # consctructing arrays for randomly choosing sample size:
    m2 = np.ones(iters, dtype=int) * 1
    # array of 2 with "iters" number of items
    m3 = np.ones(iters, dtype=int) * 5
    # array of 5 with "iters" number of items
    m4 = np.ones(iters, dtype=int) * 15
    # array of 15 with "iters" number of items
    m5 = np.ones(iters, dtype=int) * 30
    # array of 30 with "iters" number of items
    list_2 = []
    list_3 = []
    list_4 = []
    list_5 = []
    list_2.append(list(map(rnd_mean, m2)))
    list_3.append(list(map(rnd_mean, m3)))
    list_4.append(list(map(rnd_mean, m4)))
    list_5.append(list(map(rnd_mean, m5)))
    # rerun the rnd_mean function with different n inputs and append them into lists
```

```python
plt.subplots(figsize=(15, 8))

sns.histplot(data=pop_dist, kde=True)
# plot the distribution
plt.title("PDF/PMF function")
```

```python
fig, ax = plt.subplots(1, 4, figsize=(15, 8))
# plot a histograms
plt.subplot(1, 4, 1)
plt.ylabel(" ")
plt.title("n=2")
sns.histplot(data=list_2, bins=bin_num, kde=True, fill=True, legend=False)
plt.subplot(1, 4, 2)
plt.ylabel(" ")
plt.title("n=5")
sns.histplot(data=list_3, bins=bin_num, kde=True, fill=True, legend=False)
plt.subplot(1, 4, 3)
plt.ylabel(" ")
plt.title("n=15")
sns.histplot(data=list_4, bins=bin_num, kde=True, fill=True, legend=False)
plt.subplot(1, 4, 4)
plt.ylabel(" ")
plt.title("n=30")
sns.histplot(data=list_5, bins=bin_num, kde=True, fill=True, legend=False)


plt.suptitle("Demonstration of Central Limit Theorem")
# set the plot title

fig.text(0.5, 0.02, "mean", ha="center")
fig.text(0.01, 0.5, "Frequency", va="center", rotation="vertical")
plt.tight_layout()
plt.show()
```

### Poisson Distribution

```python
pop_dist = np.random.poisson(0.5, size=dist_size)
cons_array()
plt.subplots(figsize=(15, 8))

sns.histplot(data=pop_dist, kde=True)
plt.title("PDF/PMF function")
# plot the poisson distribution
fig, ax = plt.subplots(1, 4, figsize=(15, 8))
# plot a histograms
plt.subplot(1, 4, 1)
plt.ylabel(" ")
plt.title("n=1")
sns.histplot(data=list_2, bins=bin_num, kde=True, fill=True, legend=False)
plt.subplot(1, 4, 2)
plt.ylabel(" ")
plt.title("n=5")
sns.histplot(data=list_3, bins=bin_num, kde=True, fill=True, legend=False)
plt.subplot(1, 4, 3)
plt.ylabel(" ")
plt.title("n=15")
sns.histplot(data=list_4, bins=bin_num, kde=True, fill=True, legend=False)
plt.subplot(1, 4, 4)
plt.ylabel(" ")
plt.title("n=30")
sns.histplot(data=list_5, bins=bin_num, kde=True, fill=True, legend=False)


plt.suptitle("Demonstration of Central Limit Theorem")
# set the plot title

fig.text(0.5, 0.02, "mean", ha="center")
fig.text(0.01, 0.5, "Frequency", va="center", rotation="vertical")
plt.tight_layout()
plt.show()
```
