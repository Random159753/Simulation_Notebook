```python
n_iters = 10 ** 3

pr_1_s = 1
pr_2_s = 1
pr_3_s = 0
pr_4_s = 1
pr_5_s = 0
pr_6_s = 1
pr_7_s = 1
pr_8_s = 1

pr_1_initial = 250
pr_2_initial = 650
pr_3_initial = 250
pr_4_initial = 500
pr_5_initial = 700
pr_6_initial = 30
pr_7_initial = 350
pr_8_initial = 70

pr_1_p = 0.9
pr_2_p = 0.7
pr_3_p = 0.6
pr_4_p = 0.4
pr_5_p = 0.8
pr_6_p = 0.6
pr_7_p = 0.7
pr_8_p = 0.9


pr_1_min = 600
pr_2_min = 1250
pr_3_min = 500
pr_4_min = 1600
pr_5_min = 1150
pr_6_min = 150
pr_7_min = 750
pr_8_min = 220

pr_1_mode = 750
pr_2_mode = 1500
pr_3_mode = 600
pr_4_mode = 1800
pr_5_mode = 1200
pr_6_mode = 180
pr_7_mode = 900
pr_8_mode = 250

pr_1_max = 900
pr_2_max = 1600
pr_3_max = 750
pr_4_max = 1900
pr_5_max = 1400
pr_6_max = 250
pr_7_max = 1000
pr_8_max = 320


pr_1_revenue = np.random.binomial(n=1, p=pr_1_p, size=n_iters) * np.random.triangular(
    pr_1_min, pr_1_mode, pr_1_max
)
pr_2_revenue = np.random.binomial(n=1, p=pr_2_p, size=n_iters) * np.random.triangular(
    pr_2_min, pr_2_mode, pr_2_max
)
pr_3_revenue = np.random.binomial(n=1, p=pr_3_p, size=n_iters) * np.random.triangular(
    pr_3_min, pr_3_mode, pr_3_max
)
pr_4_revenue = np.random.binomial(n=1, p=pr_4_p, size=n_iters) * np.random.triangular(
    pr_4_min, pr_4_mode, pr_4_max
)
pr_5_revenue = np.random.binomial(n=1, p=pr_5_p, size=n_iters) * np.random.triangular(
    pr_5_min, pr_5_mode, pr_5_max
)
pr_6_revenue = np.random.binomial(n=1, p=pr_6_p, size=n_iters) * np.random.triangular(
    pr_6_min, pr_6_mode, pr_6_max
)
pr_7_revenue = np.random.binomial(n=1, p=pr_7_p, size=n_iters) * np.random.triangular(
    pr_7_min, pr_7_mode, pr_7_max
)
pr_8_revenue = np.random.binomial(n=1, p=pr_8_p, size=n_iters) * np.random.triangular(
    pr_8_min, pr_8_mode, pr_8_max
)

pr_1_profit = (pr_1_revenue - pr_1_initial) * pr_1_s
pr_2_profit = (pr_2_revenue - pr_2_initial) * pr_2_s
pr_3_profit = (pr_3_revenue - pr_3_initial) * pr_3_s
pr_4_profit = (pr_4_revenue - pr_4_initial) * pr_4_s
pr_5_profit = (pr_5_revenue - pr_5_initial) * pr_5_s
pr_6_profit = (pr_6_revenue - pr_6_initial) * pr_6_s
pr_7_profit = (pr_7_revenue - pr_7_initial) * pr_7_s
pr_8_profit = (pr_8_revenue - pr_8_initial) * pr_8_s

total_profit = (
    pr_1_profit
    + pr_2_profit
    + pr_3_profit
    + pr_4_profit
    + pr_5_profit
    + pr_6_profit
    + pr_7_profit
    + pr_8_profit
)
mean_profit = np.average(total_profit)
print(mean_profit)
```
